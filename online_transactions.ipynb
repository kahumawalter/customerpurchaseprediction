```python
# Preparing the dataset
x = df.iloc[:, [1, 6]].values

# Checking the shape of the dataset
print(x.shape)

# Importing necessary libraries
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Initializing the WCSS list
wcss = []

# Loop to compute WCSS for different numbers of clusters
for i in range(1, 11):
    km = KMeans(
        n_clusters=i,
        init='k-means++',
        max_iter=300,
        n_init=10,
        random_state=42,
        algorithm='elkan',
        tol=0.001
    )
    km.fit(x)
    wcss.append(km.inertia_)

# Plotting the Elbow Method graph
plt.rcParams['figure.figsize'] = (10, 6)
plt.plot(range(1, 11), wcss, color='orange',marker='o')
plt.grid()
plt.tight_layout()
plt.title('The Elbow Method', fontsize=20)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

The elbow plot shows a significant decrease in WCSS as the number of clusters increases from 1 to 3 and after that, the reduction in WCSS reduces.  

The “elbow” point appears to be 3 clusters and after this, adding more clusters will not significantly reduce the WCSS.


```python
km = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 42)
y_means = km.fit_predict(x)
plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 100, c = 'blue', label = 'Un-interested Customers')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 100, c = 'orange', label = 'General Customers')
plt.scatter(x[y_means == 2, 0], x[y_means == 2, 1], s = 100, c = 'purple', label = 'Target Customers')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s = 50, c = 'red' , label = 'centeroid')
plt.title('Administrative Duration vs Duration', fontsize = 20)
plt.grid()
plt.xlabel('Administrative Duration')
plt.ylabel('Bounce Rates')
plt.legend()
plt.show()
```

The blue dots represent un-interested customers who have low administrative duration and a wide range of bounce rates. Meaning that they spend little time on administrative pages and tend to leave quickly.

The orange dots represent general customers who tend to have moderate interaction with the administrative page and have varying bounce rates. This means that they spend more time on the administrative page than the un-interested customers but also have a tendency to leave without further engagement.

The purple dots represent target customers who have a higher administrative duration indicating more engagement with the pages and low bounce rates. This suggests that they are likely to stay and interact with the site.

The red dots are the central points of the three clusters, which are the average positions of all the points located in such clusters.


```python
# Extracting the relevant columns from the DataFrame
x = df.iloc[:, [3, 6]].values

# Initializing the list to store the Within Cluster Sum of Squares (WCSS)
wcss = []

# Loop to compute WCSS for different numbers of clusters
for i in range(1, 11):
    km = KMeans(
        n_clusters=i,
        init='k-means++',
        max_iter=300,
        n_init=10,
        random_state=0,
        algorithm='elkan',
        tol=0.001
    )
    km.fit(x)
    wcss.append(km.inertia_)

# Plotting the Elbow Method graph with orange color for the line
plt.rcParams['figure.figsize'] = (12, 6)
plt.plot(range(1, 11), wcss, color='orange', marker='o')
plt.grid()
plt.tight_layout()
plt.title('The Elbow Method', fontsize=20)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

The elbow plot shows a significant decrease in WCSS as the number of clusters increases from 1 to 2 and after that, the reduction in WCSS slows down.  

The “elbow” point appears to be 2 clusters and after this, adding more clusters will not significantly reduce the WCSS indicating diminishing returns.

```python
# one hot encoding
data1 = pd.get_dummies(df)
data1.columns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Revenue'] = le.fit_transform(df['Revenue'])
df['Revenue'].value_counts()
# getting dependent and independent variables
x=data1
# removing the target column revenue from
x = x.drop(['Revenue'], axis = 1)
y = data1['Revenue']
# checking the shapes
print("Shape of x:", x.shape)
print("Shape of y:", y.shape)
```

```python
# splitting the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
# checking the shape
print("Shape of x_train :", x_train.shape)
print("Shape of y_train :", y_train.shape)
print("Shape of x_test :", x_test.shape)
print("Shape of y_test :", y_test.shape)
```

```python
# MODELLING
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
model = RandomForestClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
# evaluating the model
print("Training Accuracy :", model.score(x_train, y_train))
print("Testing Accuracy :", model.score(x_test, y_test))
```

```python
# confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.rcParams['figure.figsize'] = (6, 6)
sns.heatmap(cm ,annot = True, cmap = 'Blues', fmt = 'd', linewidths = .5) # plotting a heatmap
plt.title('Confusion Matrix') # Settting title
plt.xlabel('Predicted')
plt.ylabel('Actual')
# classification report
cr = classification_report(y_test, y_pred)
print(cr)
```

The confusion matrix represents a negative class 0(False) and a positive class 1(True). The matrix shows that the model correctly predicted 3006 instances as false (true negative) and 317 instances as true (true positive). However, the model also incorrectly predicted 118 as true when they were false (false positives) and 258 instances as false when they were true (false negatives).

```python

# Compute the confusion matrix
cm = confusion_matrix(y, model.predict(x))

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Display the confusion matrix as an image
ax.matshow(cm, cmap='Blues')

# Disable grid
ax.grid(False)

# Set axis ticks and labels
ax.xaxis.set(ticks=(0, 1), ticklabels=('Predicted 0s', 'Predicted 1s'))
ax.yaxis.set(ticks=(0, 1), ticklabels=('Actual 0s', 'Actual 1s'))

# Annotate the matrix with text
for i in range(2):
    for j in range(2):
        ax.text(j, i, cm[i, j], ha='center', va='center', color='red')

# Show the plot
plt.show()
```

```python
from sklearn.metrics import RocCurveDisplay

# Plotting the ROC curve
RocCurveDisplay.from_estimator(model, x_test, y_test, color='orange')
plt.title('ROC Curve', fontsize=18)
plt.show()
```

The orange line represents the performance of the RandomForest classifier against the false positive rate and true positive rate.

According to the graph, the model can identify many True Positives with relatively few false positives.

With an Area Under the Curve score at 0.92, it shows that the Random Forest Classifier was good at discriminating between the True Positive and False Positive
